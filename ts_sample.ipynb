## 0.1时间序列简介

在数学上，随机过程被定义为一族时间随机变量，即$\{x(t), t \in T\}$,其中T表示时间t的变动范围。当$T = \{0, \pm 1, \pm 2, \dots \} $时，此类随机过程x(t)是离散时间t的随机函数，称为时间序列。

时间序列的构成要素有：长期趋势，季节变动，循环变动，不规则变动
1. 长期趋势(T)是在较长时期内受某种根本性因素作用而形成的总的变动趋势
2. 季节变动(S)是在一年内随着季节的变化而发生的有规律的周期性变动
3. 循环变动(C)是以若干年为周期所呈现出的波浪起伏形态的有规律的变动
4. 不规则变动(I)是一种无规律可循的变动，包括严格的随机变动和不规则的突发性影响很大的变动两种类型

本篇基于同方差的假设，介绍AR、MA、ARMA和ARIMA模型。

## 0.2序列的平稳性

时间序列的平稳性，和其它随机过程一样，分为严平稳和宽平稳。这里不给出严谨的定义，只简要阐述内容。严平稳的要求很严格，需要两个时间序列在时间的平移变换下，其联合分布保持不变。而宽平稳只需要序列的均值、方差相同即可。但是一般情况下，严平稳和宽平稳无法互相推出；若存在二阶矩过程，则可由严平稳推出宽平稳；特殊情况下（如正态分布）时，宽平稳能够推出严平稳。

由于严平稳的实现较为困难，所以一般来说平稳时间序列指的是宽平稳。如果是非平稳时间序列（如下面所展示的，上证指数一年内的收盘价），可以通过多次差分的方法处理为宽平稳序列。

## 0.3自相关函数（ACF）

相关系数衡量了两个序列的线性相关程度，而自相关函数，顾名思义就是衡量自己和自己的相关程度，即$x(t)$和过去某个时间$x(t-l)$的相关性：

$$\rho_l = \frac{Cov(x_t, x_{t-l})}{\sqrt{Var(x_t)Var(x_{t-l})}} = \frac{Cov(x_t, x_{t-l})}{Var(x_t)}$$

下面进行统计检验，原假设为所有自相关系数为0，检验统计量为

$$Q(m) = T(T+2)\sum_{l=1}^{m} \frac{\hat{\rho}_l^2}{T-l}$$

服从自由度为m的卡方分布。取显著水平为0.05，若所有p值均小于0.05则拒绝原假设，认为该序列存在序列相关性。

例如，取上证指数2016年全年的收盘价，计算自相关函数，检验统计量Q以及p值。可以初步判断，上证指数去年的收盘价不存在显著序列相关性。

```python
closedata = IndexData['close'] # 上证指数每日收盘价
m = 10 # 检验10个自相关系数

acf,q,p = sm.tsa.acf(closedata, nlags=m, qstat = True)
out = np.c_[range(1,11), acf[1:], q, p]
output = pd.DataFrame(out, columns=['lag', 'ACF', 'Q', 'P-value'])
output = output.set_index('lag')
output
```

再例如，取上证指数每日涨跌幅，进行同样计算发现，也不存在显著序列相关性。

```python
changedata = IndexData['changepct'] # 上证指数每日涨跌幅
m = 10 # 检验10个自相关系数

acf,q,p = sm.tsa.acf(changedata,nlags=m,qstat=True)  ## 计算自相关系数 及p-value
out = np.c_[range(1,11), acf[1:], q, p]
output=pd.DataFrame(out, columns=['lag', "AC", "Q", "P-value"])
output = output.set_index('lag')
output
```

# 1.自回归（Autoregressive，AR）模型

首先介绍一阶自回归模型，即AR(1)，也就是运用t-1时刻的值预测t时刻的值，即
$$r_t = \phi_0 + \phi_1 r_{t-1} + \epsilon_t$$
其中最后一项代表白噪声序列（随机过程的任意两项协方差为0，且均值和方差均为常数的序列），由一阶自回归模型可以推广到p阶：
$$r_t = \phi_0 + \phi_1 r_{t-1} + \dots + \phi_p r_{t-p} + \epsilon_t$$

假定序列是弱平稳的，利用平稳序列和白噪声序列的均值方差性质，可以得出：
$$\mu = \frac{\phi_0}{1 - \phi_1 - \phi_2 - \dots - \phi_p}$$
其中，$\mu$是{rt}的均值。

当p阶特征方程
$$1 - \phi_1 x - \phi_2 x^{2} - \dots - \phi_p x^{p}  = 0 $$
的特征根落在单位圆之外，也就是特征根的倒数落在单位圆之内时，该p阶自回归序列是平稳的。下面运用此方法计算上证指数2016年每日收益率序列的平稳性，通过特征根倒数在单位圆的分布情况，可以看出该序列是平稳的。

```python
changedata = IndexData['changepct'] # 载入收益率序列
changedata = np.array(changedata, dtype=np.float)
model = sm.tsa.AR(changedata)
results_AR = model.fit()
plt.figure(figsize=(10,4))
plt.plot(changedata, 'b', label = 'changepct')
plt.plot(results_AR.fittedvalues, 'r', label = 'AR model')
plt.legend()
```

```python
print len(results_AR.roots) # 模型阶数
```

```python
r1 = 1
theta = np.linspace(0,2*np.pi,360)
x1 = r1*np.cos(theta)
y1 = r1*np.sin(theta)
plt.figure(figsize=(6,6))
plt.plot(x1,y1,'k')  # 画单位圆
roots = 1/results_AR.roots  # 计算results_AR.roots后取倒数
for i in range(len(roots)):
    plt.plot(roots[i].real,roots[i].imag,'.r',markersize=8)  #画特征根
plt.show()
```

## 1.1确定AR模型的阶数

对于线性回归模型的阶数，一般认为随着阶数的升高，会提高模型的精度。但是过高的阶数会造成模型过于复杂。下面介绍两种确定AR模型阶数的方法：偏相关函数（pACF）和信息准则函数（AIC、BIC、HQIC）.

### 1.1.1偏相关函数(pACF)
首先引入阶数依次升高的AR模型：
$$\begin{eqnarray}
&&r_t = \phi_{0,1} + \phi_{1,1}r_{t-1} + \epsilon_{1t} \\
&&r_t = \phi_{0,2} + \phi_{1,2}r_{t-1} + \phi_{2,2}r_{t-2} + \epsilon_{2t} \\
&&r_t = \phi_{0,3} + \phi_{1,3}r_{t-1} + \phi_{2,3}r_{t-2} + \phi_{3,3}r_{t-3} + \epsilon_{3t} \\
&&r_t = \phi_{0,4} + \phi_{1,4}r_{t-1} + \phi_{2,4}r_{t-2} + \phi_{3,4}r_{t-3} + \phi_{4,4}r_{t-4} + \epsilon_{4t} \\
&&\dots \dots \dots \dots
  \end{eqnarray}$$
等式右边第一项为常数项，最后一项为误差项，$\hat{\phi}_{i,i}$为{rt}的间隔为i的样本偏自相关函数。$\hat{\phi}_{i,i}$表示的是在第i-1行的基础上添加的一个阶数对{rt}的贡献。对于AR模型，当超过某个值p时，偏自相关函数应逐渐趋近于0，这既符合我们的直观认识，也有相应的理论证明可以找到。

下面对2016年上证指数日涨跌幅作出偏自相关函数图，在5%的显著水平下，取置信区间为$\frac{2}{\sqrt{n}} = \frac{2}{\sqrt{243}} \approx 0.128$

```python
fig = plt.figure(figsize=(20,5))
ax1 = fig.add_subplot(111)
fig = sm.graphics.tsa.plot_pacf(changedata,ax=ax1) #运行速度很慢, 建议耐心等待...
```

很遗憾，从上图中并不太容易确定一个明显的阶数p。不过确实可以看出，随着时间的增加，偏相关函数逐渐趋近于0. 这个结果给了我们两点思路：首先偏相关函数趋近于0，说明运用AR模型模拟该序列是有一定的合理性；其次上图很难确定阶数p或阶数p很大，说明模型可能有MA（滑动平均，将在后面介绍）的部分。它至少给了我们一些感性的认识。

### 1.1.2信息准则函数

目前比较常用的信息准则函数包括赤池信息量（AIC），贝叶斯信息量（BIC）和汉南-奎因信息量（HQIC）。他们都基于似然函数L和参数的数量k，其中n表示样本大小。由于Python中有相应的函数，所以公式仅作了解：
$$\begin{eqnarray}
&&AIC = -2 \ln L + 2 k \\
&&BIC = -2 \ln L + \ln n \cdot k \\
&&HQIC = -2 \ln L + \ln(\ln n) \cdot k
\end{eqnarray}$$

我们的目标是，选择阶数k使得信息量的值最小。

```python
aicList = []
bicList = []
hqicList = []
for i in range(1,8):  #从1阶开始算
    order = (i,0)  # 使用ARMA模型，其中MA阶数为0，只考虑了AR阶数。
    chgdataModel = sm.tsa.ARMA(changedata,order).fit()
    aicList.append(chgdataModel.aic)
    bicList.append(chgdataModel.bic)
    hqicList.append(chgdataModel.hqic)
    print i
```

```python
plt.figure(figsize=(10,6))
plt.plot(aicList,'ro--',label='aic value')
plt.plot(bicList,'bo--',label='bic value')
plt.plot(hqicList,'ko--',label='hqic value')
plt.legend(loc=0)
```

计算信息准则函数的时候，我们发现随着阶数的增加，运算速度越来越慢，所以只取了前7阶，可以看出AIC给出的阶数为5，BIC和HQIC给出的阶数为1. 当然，前面提到只用AR模型预测该序列是不准确的，且只计算前7阶的结果也不足以判断最佳阶数。

# 2.滑动平均（moving-average, MA）模型

除了简单性以外，没有特别的理由假定AR模型的阶数是有限的，考虑无穷阶AR模型为:

$$r_t = \phi_0 + \phi_1 r_{t-1} + \phi_2 r_{t-2} + \dots + \epsilon_t$$

然而这样的AR模型实际意义较小，我们也通过上面的偏自相关函数看到，AR模型的阶数可以设定的很大，导致模型过于复杂。我们现在引入最简单的MA模型，即MA(1)，假设上述AR模型中的$\phi_i$满足某种限制：

$$r_t = \phi_0 - \theta_1 r_{t-1} - \theta^2_1 r_{t-2} - \theta^3_1 r_{t-3} - \dots + \epsilon_t$$

错位相减之后得到MA(1)模型的一般形式：

$$r_t = c_0 + \epsilon_t - \theta_1 \epsilon_{t-1}$$

推广到q阶MA模型就是：

$$r_t = c_0 + \epsilon_t - \theta_1 \epsilon_{t-1} - \dots - \theta_q \epsilon_{t-q}$$

回到文章最初提到的平稳性定义，由于MA模型是白噪声序列（均值方差恒定）的有限线性组合，因此MA模型具有弱平稳性。这告诉我们MA模型本身就具有平稳性，并不需要像AR模型一样通过特征根的方法判断其平稳性。

同时，滑动平均顾名思义就是在邻近区域内做平均并不断移动，因此可以看出（也可证明）MA(q)序列之和其前q个值自相关，是一个“有限记忆”模型。

## 2.1确定MA模型的阶数

之前我们知道，通过计算偏自相关函数（pACF）可以粗略确定AR模型的阶数。而对于MA模型，需要采用自相关函数（ACF）进行定阶。

```python
fig2 = plt.figure(figsize=(20,5))
ax2=fig2.add_subplot(111)
fig2 = sm.graphics.tsa.plot_acf(changedata,ax=ax2)
```

我们发现自相关函数在16阶之后，取值就处于置信区间内，并逐渐趋于0，可粗略判定该序列MA模型为16阶。由于模型阶次过高，计算量太大，在此以10阶模型为例，作出MA模型与原日涨跌幅的图像。

```python
results_MA = sm.tsa.ARMA(changedata, (0,10)).fit()
plt.figure(figsize=(10,4))
plt.plot(changedata, 'b', label = 'changepct')
plt.plot(results_MA.fittedvalues, 'r', label = 'MA model')
plt.legend() # 运行时间较长
```

# 3.自回归-滑动平均（ARMA）模型

大多数情况下，AR和MA模型都很难单独地描述和预测一个序列，这时我们将它们合在一起，使得所用的参数个数保持很小，预测效果也有一定的改善。首先引入ARMA(1,1)模型：

$$r_t - \phi_1 r_{t-1} = \phi_0 + \epsilon_t - \theta_1 \epsilon_{t-1}$$

通常要求$\phi_1 \neq \theta_1$.

推广到ARMA(p,q)，p是AR模型的阶数，q是MA模型的阶数，得到：

$$r_t = \phi_0 + \epsilon_t + \sum^{p}_{i=1} \phi_i r_{t-i} + \sum^{q}_{i=1} \theta_i \epsilon_{t-i}$$

由于误差项是白噪声序列（期望为0），ARMA模型求得的期望和AR模型一样，同样判断特征根的大小确定其平稳性。

## 3.1确定ARMA模型的阶数

我们可以通过pACF和ACF分别确定AR和MA模型的阶数，或者查找EACF表来确定（在《金融时间序列分析（第二版）》人民邮电出版社 第2.6节有相应介绍）。这里还是倾向于使用信息准则定阶方法，即AR模型中提到的AIC，BIC和HQIC等。为了缩短运行时间，设置AR模型最大阶数为6，MA模型最大阶数为4。即使这样，运算时间仍然很长，来看结果：

```python
import warnings
warnings.filterwarnings("ignore")
sm.tsa.arma_order_select_ic(changedata,max_ar=6,max_ma=4,ic='aic')['aic_min_order']  # AIC
```

```python
sm.tsa.arma_order_select_ic(changedata,max_ar=6,max_ma=4,ic='bic')['bic_min_order']  # BIC
```

```python
sm.tsa.arma_order_select_ic(changedata,max_ar=6,max_ma=4,ic='hqic')['hqic_min_order'] # HQIC
```

我们以AIC准则为准，即使用ARMA(6,2)模型进行建模分析。

## 3.2ARMA模型的预测和检验

根据信息准则定阶方法，建立ARMA(6,2)模型，仍然对上证指数日涨跌幅进行预测，并进行同样的拟合优度检验。

```python
train = changedata[:-10]
test = changedata[-10:]
chgdataModel_ARMA = sm.tsa.ARMA(train, (6,2)).fit()
```

```python
predicts_ARMA = chgdataModel_ARMA.predict(233, 242, dynamic = True)
print len(predicts_ARMA)
compare_ARMA = pd.DataFrame()
compare_ARMA['original'] = test
compare_ARMA['predict_ARMA'] = predicts_ARMA
compare_ARMA
```

```python
compare_ARMA.plot()
```

```python
delta_ARMA = chgdataModel_ARMA.fittedvalues - train # 拟合优度检验
score_ARMA = 1 - delta_ARMA.var()/train.var()
print score_ARMA
```

预测结果的拟合优度得分仍然很低，但是从预测图上来看，好像比之前强了一点，至少在中间部分效果还不错。

可能的原因有：
1. 信息准则定阶时，为减少计算量，我们设定了AR和MA模型的最大阶数，这样得到的阶数只是局部最优解。
2. ARMA模型不足以预测整个上证指数的变化趋势（这也正常，不然人人都可以赚大钱了）。

# 4.差分自回归滑动平均（ARIMA）模型

ARMA模型研究的对象为平稳时间序列。如果序列是非平稳的，需要通过差分的方法将其变为平稳时间序列，也就是所说的ARIMA模型。ARIMA模型需要确定三个阶数，AR模型的阶数p，差分的阶数d和MA模型的阶数q，通常写作ARIMA(p,d,q)。

## 4.1ARIMA模型平稳性检验

为了使结果更加明显，我们选取2016年上证指数收盘价序列进行研究。首先作出图像，直观上看它是不平稳的。

```python
closedata = IndexData['close'] # 取得上证指数2016年收盘价
closedata.plot(figsize = (15,5))
```

我们依次对收盘价、收盘价的一阶差分、收盘价的二阶差分进行ADF(Augment Dickey-Fuller)检验，具体原理在此省略。当p值小于0.05时，可以判断序列是平稳的。

```python
closedata = np.array(closedata) # 收盘价的ADF检验
adftest = sm.tsa.stattools.adfuller(closedata)
```

```python
dif1closedata = IndexData['closeDif_1'] #收盘价一阶差分的ADF检验
dif1closedata = dif1closedata[1:] # 去除第一行的NaN值
dif1closedata = np.array(dif1closedata)
adftestdif1 = sm.tsa.stattools.adfuller(dif1closedata)
```

```python
dif2closedata = IndexData['closeDif_2'] #收盘价二阶差分的ADF检验
dif2closedata = dif2closedata[2:] # 去除前两行的NaN值
dif2closedata = np.array(dif2closedata)
adftestdif2 = sm.tsa.stattools.adfuller(dif2closedata)
```

```python
output=pd.DataFrame(index=["p-value"],columns=["original", "dif1", "dif2"])
output.loc["p-value"]["original", "dif1", "dif2"] = adftest[1], adftestdif1[1], adftestdif2[1]
output
```

从p值的情况可以看出，一阶差分之后就变为了平稳序列，还可以看出差分可以使得非平稳序列快速变为平稳序列。事实上，一般而言两阶差分就足够将非平稳序列变得平稳。

这样，我们确定了d的阶数为1，下面确定AR和MA模型的阶数。

## 4.2确定ARIMA模型的阶数

和之前一样，通过pACF和ACF可以粗略判断AR和MA的阶数。这里我们仍然使用信息准则中的AIC定阶方法。

```python
sm.tsa.arma_order_select_ic(dif1closedata,max_ar=6,max_ma=4,ic='aic')['aic_min_order']  # AIC 运行速度较慢...
```
